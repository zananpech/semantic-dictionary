```python
import pandas as pd
```

```python
df = pd.read_csv('data/dict.csv')
df.head()
```

```python
df['word + definition'] = df['word'] + ' : ' + df['definition']
df['word + definition'] = df['word + definition'].astype(str)
df.head()
```

```python
print(df['word'].unique().shape)
print(df.shape)
```

```python
import chromadb
chroma_client = chromadb.Client()
```

```python
chroma_client.delete_collection('english_dictionary')
collection = chroma_client.create_collection(name="english_dictionary")
```

```python
for doc, id_ in zip(df['word + definition'].tolist(), list(map(str, df.index.tolist()))):
    collection.upsert(
        documents = [doc],
        ids = [id_]
    )
    print(f'{id_} inserted.')
```
