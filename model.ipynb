```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv('data/dict.csv')
df.head()
```

```python
feature, target = df.definition, df.word
```

```python
feature, target = df.definition, df.word
feature.shape
```

```python
vectorizer = TfidfVectorizer()
feature_enc = vectorizer.fit_transform(feature)
feature_enc
```

```python
model = MultinomialNB()
model.fit(feature_enc, target)
```

```python
model.score(feature_enc, target)
```

```python
probs = model.predict_proba(vectorizer.transform(['I have the power the do something']))[0]
np.argsort(probs)[::-1][:5]
```
